```python
from google.colab import files
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Unggah gambar KTP/SIM
uploaded = files.upload()

# Baca gambar menggunakan OpenCV
for filename, content in uploaded.items():
    image = cv2.imdecode(np.frombuffer(content, np.uint8), -1)

     # Tentukan koordinat dan ukuran kotak pertama untuk menutup agama
    x1, y1, x2, y2 = 120, 250, 250, 280  # Ganti dengan koordinat yang sesuai
    color1 = (255, 255, 255)  # Warna putih

    # Tentukan koordinat dan ukuran kotak kedua untuk menutup status kawin
    x3, y3, x4, y4 = 130, 280, 260, 325  # Ganti dengan koordinat yang sesuai
    color2 = (0, 0, 0)  # Warna hitam

    # Salin area yang akan ditutup menggunakan dua kotak warna yang berbeda
    image[y1:y2, x1:x2] = color1
    image[y3:y4, x3:x4] = color2

    # Tampilkan gambar yang telah dimodifikasi
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```
