Identifikasi kenapa menggunakan konversi warna BGR2RGB pada matplotlib dan tanpa
konversi pada opencv

Jawab:

Konversi warna dari BGR ke RGB diperlukan saat menggunakan Matplotlib karena Matplotlib mengharapkan format RGB (Red, Green, Blue) untuk tampilan gambar, sementara OpenCV secara alami menggunakan format BGR (Blue, Green, Red) untuk merepresentasikan gambar. Oleh karena itu, kita perlu mengonversi gambar dari BGR ke RGB sebelum menampilkannya di Matplotlib agar warnanya tampil dengan benar. Pada OpenCV, kita tidak perlu melakukan konversi karena format BGR sudah sesuai dengan representasi gambar yang digunakan oleh OpenCV.

```python
from google.colab import files
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Unggah gambar KTP/SIM
uploaded = files.upload()

# Baca gambar menggunakan OpenCV
for filename, content in uploaded.items():
    image = cv2.imdecode(np.frombuffer(content, np.uint8), -1)

    # Tentukan koordinat dan ukuran kotak pertama untuk menutup agama
    x1, y1, x2, y2 = 190, 278, 258, 301  # Ganti dengan koordinat yang sesuai
    color1 = (0, 255, 255)  # Warna kuning

    # Tentukan koordinat dan ukuran kotak kedua untuk menutup status kawin
    x3, y3, x4, y4 = 190, 302, 264, 324  # Ganti dengan koordinat yang sesuai
    color2 = (0, 0, 0)  # Warna hitam

    # Salin area yang akan ditutup menggunakan dua kotak warna yang berbeda
    image[y1:y2, x1:x2] = color1
    image[y3:y4, x3:x4] = color2

    # Tampilkan gambar yang telah dimodifikasi
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```
